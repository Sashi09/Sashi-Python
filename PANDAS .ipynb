```python
df1 = pd.DataFrame({'key': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="key")
```

```python
#merge adds two data frames based on the column name
#here in above example
#ko is available in both df1 and df2 so it is merged
#k8 is in df1 but not in df2 where we have k1...so this row is not merged
```

```python
#it is going to perfoem inner merge
```

```python
pd.merge(df1,df2,on="key" ,how = 'left')
```

```python
#here it will consider left key (df1) first and merge the data
```

```python
#as 'k8' is not available in right data it will give value as nan
```

```python
#in similar way we can do operations from right
```

```python
pd.merge(df1,df2,on="key" ,how = 'right')
```
